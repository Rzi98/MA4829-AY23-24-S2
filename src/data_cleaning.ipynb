```python
import pandas as pd

df = pd.read_excel(io='../data/SURVEY RESULTS_2024.xlsx', sheet_name='Customer preference in car')

df.head(3)
```

### **Renaming columns to shorten names** ###

```python
df.rename(columns={
    'Which age group do you belong to?': 'Age group',
    'What is your gender?': 'Gender',
    'Which category do you currently belong to?': 'Category',
    'Which of the following best describes you?': 'Marital status',
    'Which of these factors are important to you when deciding which car to purchase?': 'Deciding factors to buy car',
    'How likely are you to opt for customised vehicle if there were no extra charges? ': 'Likelihood of customised vehicle if FOC',
    'Which of the following exterior components would you choose to customise (texture, layout, size, etc)? ': 'Exterior components to customise',
    'Which of the following interior components would you choose to customise (texture, layout, size, etc)? ': 'Interior components to customise',
    'How much are you willing to spend on car customisation if surcharges are applicable?': 'Customisation Budget',
    'How much are you willing to pay for the personalised design? ': 'Personalisation Budget',
    'Are you interested in designing your own components to personalise your car? ': 'Interest in personalising car',
    'Do you have any 3D design experience that would help with the design process? (e.g. AutoCAD, SolidWorks, Blender, etc)': 'Any cadding experience',
    'Please give us any design ideas to make the car uniquely Singaporean.': 'Singaporean design idea feedback',
}, inplace=True)

df.head(3)
```

### **Giving Unique Customer ID to each row** ###

- In a real world database, each customer will be given a unique ID for tracking & analysis
- But in this case, since it is a survey, it is not really important to know who the customer is but I will just add it

```python
import uuid

df['Respondent ID'] = [str(uuid.uuid4()) for _ in range(len(df))]
df = df[['Respondent ID'] + [col for col in df.columns if col != 'Respondent ID']]
df.head()
```

```python
df.iloc[:, 1:].describe()
```

## **Replace NaN with '-'** ##

```python
df.fillna('-', inplace=True)
df.iloc[:, 1:].describe()
```

### **Replace 100-500 to under 500 for customisation budget** ###

- only 1 row value to simplify categorisation
- Adding ordinal columns for customisation budget and personalisation budget

```python
df['Customisation Budget'] = df['Customisation Budget'].replace('100-500', 'under 500')
df['Customisation Budget'].value_counts()
```

```python
budget_mapping = {0: 1, 'under 500': 2, '500-1000': 3, 'over 1000': 4}

df['Customisation Budget Ordinal'] = df['Customisation Budget'].map(budget_mapping)
df['Personalisation Budget Ordinal'] = df['Personalisation Budget'].map(budget_mapping)
df.head()
```

```python
df.to_csv('../data/cleaned_data.csv', index=False)
df.to_excel('../data/cleaned_data.xlsx', index=False)
```
