```python
import pandas as pd

df = pd.read_csv('../data/cleaned_data.csv')

df.head(2)
```

```python
df = df[['Interest in personalising car', 'Any cadding experience']]
df.head()
```

## **Interest in Personalising Car**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 7))
ax = sns.countplot(df, x="Interest in personalising car", hue="Any cadding experience")

for p in ax.patches:
    count = int(p.get_height())
    if count > 0:
        ax.annotate(f'{count}', (p.get_x() + p.get_width() / 2., count),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

## **Customisation Budget** ##

```python
import pandas as pd

df = pd.read_csv('../data/cleaned_data.csv')

df = df[['Customisation Budget', 'Personalisation Budget']]

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

custom_order = ["0", "under 500", "500-1000", "over 1000"]

plt.figure(figsize=(8, 7))
ax = sns.countplot(df, x="Customisation Budget", order=custom_order)

for p in ax.patches:
    count = int(p.get_height())
    if count > 0:
        ax.annotate(f'{count}', (p.get_x() + p.get_width() / 2., count),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/cleaned_data.csv')

df = df[['Customisation Budget', 'Personalisation Budget']]
order = ['Low', 'Mid Low', 'Mid High', 'High']

label_mapping = {'0': 'Low', 'under 500': 'Mid Low', '500-1000': 'Mid High', 'over 1000': 'High'}
df['Customisation Budget'] = df['Customisation Budget'].map(label_mapping)

plt.figure(figsize=(8, 7))
ax = sns.countplot(df, x="Customisation Budget", order=order)

for p in ax.patches:
    count = int(p.get_height())
    if count > 0:
        ax.annotate(f'{count}', (p.get_x() + p.get_width() / 2., count),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```
